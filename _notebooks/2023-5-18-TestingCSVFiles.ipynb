```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

stat = input("What statistic would you like to analyze today? Games Played, Points, Assists, Rebounds, Steals, Blocks, Field Goal %, or 3 Point %?")

plt.title(str(stat) + " of each Atlanta Hawks Player")

if stat == "Games Played":
    stat = 1
elif stat == "Points":
    stat = 2
elif stat == "Assists":
    stat = 3
elif stat == "Rebounds":
    stat = 4
elif stat == "Steals":
    stat = 5
elif stat == "Blocks":
    stat = 6
elif stat == "Field Goal %":
    stat = 7
elif stat == "3 Point %":
    stat = 8

data = pd.read_csv('files/hawks.csv')
x_pos = np.arange(len(data))
df = pd.DataFrame(data)

X = list(df.iloc[:, 0])
Y = list(df.iloc[:, stat])

# Plot the data using bar() method
plt.bar(X, Y, color='g')
plt.xlabel("Players")
plt.ylabel("Statistics")
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
